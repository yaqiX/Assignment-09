## Make a Python string that contains the text of a Shakespeare play 

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
with open('romeo and juliet.txt','r') as f:
    text1 = f.readlines()
```

```python
text1
```

```python
type(text1)
```

```python
text=' '.join(text1)
```

```python
type(text)
```

```python
text
```

## Tokenization

```python
sent = sent_tokenize(text)
```

```python
print(word_tokenize(sent[1]))
```

```python
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
words
```

## Remove Stopwords

```python
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
print(punctuation)
```

```python
print(stopwords.words('english'))
```

```python
myStopWords = list(punctuation) + stopwords.words('english')
```

```python
text_ns = []
for i in words:
    if i.lower() not in myStopWords:
        text_ns.append(i)
print(text_ns)
# text_ns: text with no stopwords
```

## Find the top 20 most frequent words in the play

```python
word_list = text.split()
```

```python
wordfreq = [word_list.count(w) for w in text_ns] 
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(text_ns)
```

```python
freq
```

```python
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

```python
romnjulstopwords=["'s","thou","'d","rom","thy","thee","jul","shall","'ll","project","enter","exit","may","hath","gutenberg-tm","would","one","must","upon","yet","much","doth"]
```

```python
StopWords_romnjul = list(punctuation) + stopwords.words('english') + romnjulstopwords
```
